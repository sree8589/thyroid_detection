Hyperthyroid (A), T3 toxic (B), Toxic goitre (C), and Secondary toxic (D) are different Hyperthyroid conditions.

Hypothyroid (E), Primary Hypothyroid (F), Compensated Hypothyroid (G), and Secondary Hypothyroid (H) are different Hypothyroid conditions.

"-" indicates no Thyroid conditions.

All Anti-thyroid treatments listed in this dataset, such as Antithyroid drugs (O), I131 treatment (P) and Surgery (Q) are recommended in case of Hyperthyroid Disorders.

Elevated thyroid hormones (T), also known as hyperthyroidism, occurs when the thyroid gland produces too many thyroid hormones.

Elevated Thyroxine-Binding Globulin (TBG) (S) is directly assocaited with Hyperthyroidism.

```python
hyperthyroid_conditions = ['A', 'B', 'C', 'D', 'O', 'P', 'Q', 'S', 'T']
hypothyroid_conditions = ['E', 'F', 'G', 'H']
normal_conditions = ['-']

def categorize_target(value):
    diagnoses = value.split('|')
    for diagnosis in diagnoses:
        if diagnosis in hyperthyroid_conditions:
            return 'Hyperthyroid'
    for diagnosis in diagnoses:
        if diagnosis in hypothyroid_conditions:
            return 'Hypothyroid'
            
    for diagnosis in diagnoses:
        if diagnosis in normal_conditions:
            return 'Negative'


data_copy['target'] = data_copy['target'].apply(categorize_target)
```

### XG boost with oversampling

```python
pip install imblearn
```

The dataset used for thyroid detection had a significant class imbalance
Such an imbalance can lead to models predicting the majority class (Negative) more often, neglecting minority classes (Hypothyroid and Hyperthyroid).
Oversampling using SMOTE (Synthetic Minority Over-sampling Technique) was applied to generate synthetic data points for minority classes. This balanced the class distribution and improved model learning.

```python
import joblib
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.utils.class_weight import compute_sample_weight
smote = SMOTE(random_state=42)
xtrain_resampled, ytrain_resampled = smote.fit_resample(xtrain, ytrain_xg)
sample_weights = compute_sample_weight(class_weight='balanced', y=ytrain_resampled)
model_xg_os = XGBClassifier(random_state=10)
model_xg_os.fit(xtrain_resampled, ytrain_resampled, sample_weight=sample_weights)
joblib.dump(model_xg_os, 'xgb_model.pkl')
```

```python
xg_os_predict=model_xg.predict(xtest)
print(model_xg_os.score(xtrain,ytrain_xg))
print(model_xg_os.score(xtest,ytest_xg))
```

```python
print(precision_score(ytest_xg,xg_os_predict,average='micro'))
print(precision_score(ytrain_xg,model_xg_os.predict(xtrain),average='micro'))
```

```python
print(recall_score(ytest_xg,xg_os_predict,average='micro'))
print(recall_score(ytrain_xg,model_xg_os.predict(xtrain),average='micro'))
```

XGBoost with oversampling maintained excellent test performance.

There was no major overfitting, and the recall remained high.

Oversampling further balanced the class predictions, making it a robust model for medical diagnosis.
